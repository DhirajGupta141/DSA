### Dict Creation in Python 

```python
d = dict() 
print(d)
```

```python
d = {}
print(d)
```

### d.items()

```python
d = {1:'a', 2:'b'}

for key, value in d.items():
    print(f"For the key : {key} value is {value}")
```

### list(d.items())

```python
d = {28:[1,23,4], 20 : [4,5,10,1]}
```

```python
d.items()
```

```python
list(d.items())
```

```python
for sum, elements in list(d.items()):
    print(f"For the sum : {sum} elements are {elements}")
```

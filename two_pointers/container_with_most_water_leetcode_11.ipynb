#### 11. Container With Most Water

```python
from typing import List 

class Solution:
    def maxArea(self, a: List[int]) -> int:

        n = len(a)

        i = 0 
        j = n-1 

        maxi = 0

        while i < j:
            maxi = max(maxi , (j - i) * min(a[i],a[j]))
            
            if a[i] < a[j]:
                i += 1   # we are greedy.. so we will increase the pointers with the less value 
            else:
                j -= 1
        
        return maxi 


sol = Solution()
a = [1,8,6,2,5,4,8,3,7]
ans = sol.maxArea(a)

print(f"Maximum Water which can be accumulated in the height array : {a} is {ans}") 

        
```

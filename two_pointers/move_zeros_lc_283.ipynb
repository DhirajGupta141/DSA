### Leetcode : 283 , Move Zeros

Given an integer array nums, move all 0's to the end of it while maintaining the relative order of the non-zero elements.

Note that you must do this in-place without making a copy of the array.

 

Example 1:

Input: nums = [0,1,0,3,12]
Output: [1,3,12,0,0]
Example 2:

Input: nums = [0]
Output: [0]

```python

"""
This Questions is of two pointers because .. here we need 2 things 
1. First we need to track the nonzeros values 
2. Second we need to track the Zero Values
"""

class Solution:
    def moveZeroes(self, a: list[int]) -> None:

        n = len(a)

        i = 0  # Non-Zeros values 
        j = 0  # need all zeros to be at the end  

        while j < n:
            if a[j] != 0:
                a[j] , a[i] = a[i] , a[j]
                i += 1
            
            j += 1
        
        
        
# Note : this is inpce Solution So cannot See here !
        
```

#### How to create a singly LinkededList

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None 
```

```python
class LinkedList:

    def __init__(self):
        self.head = None   # This will always  be at the first node.. for Insert .. we are using a supporting curr variable.
    
    def insert_at_head(self, value):
        newNode = Node(value)
        newNode.next = self.head   # new node would be the first node of the LINKEDLIST 
        self.head = newNode

        return 

    def insert_at_end(self,value):
        newNode = Node(value)

        # Its first element 
        if not self.head:
            self.head = newNode
            return 

        # if we already have elements in the Linkeded list self.head is not None 
        curr = self.head
        while curr.next:
            curr = curr.next
        
        curr.next = newNode

        return 
    
    def print_linked_list(self):
        curr = self.head
        while curr:
            print(curr.data , end= ' -> ')
            curr = curr.next 
        
        print('None')


## execution 
ll = LinkedList()
ll.insert_at_end(4)
ll.insert_at_end(10)
ll.insert_at_head(-1)
ll.insert_at_head(-100)


ll.print_linked_list()



    
    


```

### Add Two Number Leetcode LinkedList : 2 

Carry = x // 10 

remainder = x % 10 


```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        """
        This function takes two linked lists representing two non-negative integers,
        where the digits are stored in reverse order. 
        It adds the two numbers and returns the sum as a linked list in the same reverse order format.
        """
        dummy = ListNode()
        current = dummy # helper variable 
        carry = 0 

        while l1 or l2 or carry:  # loop until both lists are fully traversed and no carry is left
            a = l1.val if l1  else 0 
            b = l2.val if l2 else 0 
            total = a + b + carry

            new_node_val = total % 10 
            carry = total // 10

            current.next = ListNode(new_node_val)

            if l1:
                l1 = l1.next 
            if l2:
                l2 = l2.next 
            
            current = current.next 
    
        return dummy.next 

```

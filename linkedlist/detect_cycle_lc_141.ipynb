##### Leetcode : 141 : Detect Cycle in the LinkedList 

```python
class Solution:
    def hasCycle(self, head) -> bool:

        if not head:
            return False 
        
        if not head.next:
            return False 

        slow = head
        fast = head 
        
        ## Slow check is not needed as fast is ruuning 2 points at once 
        while fast and fast.next:  # as we are doing fast.next(fast should not be None ) and then fast.next.next(fast.next should not be None )
            slow = slow.next 
            fast = fast.next.next 

            if slow == fast:
                return True 
        
        return False 
```

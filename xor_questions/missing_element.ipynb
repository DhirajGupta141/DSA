#### XOR : Cancel out the same elements :  1^1 = 0 

#### Finding Missing Number 

a : [1,2,4,5]

output : 3

#### Using Formula 

```python
def findMissing(a,k):
    try:
        n = len(a)
        sumi = sum(a)
        sumi_k = k*(k+1) / 2
        missing_element = int(sumi_k-sumi)
        print(f"Sum of n elements is : {sumi_k} and sum of array elements is : {sumi}")
        print(f"Missing element is : {missing_element}")

        return missing_element


    except Exception as e:
        raise e 

a = [1,2,4,5]
k = 5
missing_element = findMissing(a, k)
print(f"Missing Element in the array is {a} is : {missing_element}")
```

#### Using XoR Operation : XoR of the same element i 0 

--> 1^1^2^2^3^4^4^5^5 = 3

```python
def xorMissing(a,k):
    try: 
        xor1=0
        xor2= 0
        for i in range(1,k+1):
            xor1 = xor1^i
        
        for i in a:
            xor2 = xor2^i 
        
        return xor1^xor2
    except Exception as e:
        raise e 
       
a = [1,2,4,5]
k = 5
missing_element = xorMissing(a, k)
print(f"Missing Element in the array is {a} is : {missing_element}") 
```

#### Longest Subarray with 0 Sum : Same concept as : Longest Subarray with sum K 


#### Intitution 

0. Sliding window techmique yaha nhi chalega because .. SW works only for the positive array values as we have condn if sum > k : reduce and this is not there for the negative value 
1. running sum : prefix sum ham nikalte jayenge and har ek prefix sum and uska kon se index pr kitna prefix sum aaya wo store krenge in the dict prefix_dict = {sum : index}
2. Aage chal kr lets say at index j hame same prefix sum value aa gya jo hamne past me ith index pr aaya tha .. iska matlab .. ith to jth index k bich ke elements ka sum 0 hai .. 

```python
def longestSubarraySumZero(a):
    try:
        prefixSumDict = {0:-1} # {sum : index} 0 prefix sum dekha tha hamne at index -1  
        n = len(a)
        maxi = 0

        prefixSum = 0
        for j , num in enumerate(a):
            prefixSum += num

            if prefixSum in prefixSumDict:
                i = prefixSumDict[prefixSum]
                maxi = max( maxi , j-i)
            else:
                prefixSumDict[prefixSum] = j
        
        return maxi


    except Exception as e:
        raise e
```

```python
if __name__ == '__main__':
    a = [15,-2,2,-8,1,7,10]
    ans = longestSubarraySumZero(a)
    print(f"Longest Subarray with sum 0 for the array {a} is {ans}")
```

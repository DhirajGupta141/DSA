#### 560. Subarray Sum Equals K : Same as Longest Subarray with Sum 0 : Here k = 0

SLIDING WINDOW Doesnt work here because of the Negative values in the Array : as we have only the condition of if sum > k : reduce 

We need to use the prefix sum Approach for the optimized solution and the bruteforce we can do by running the 2 loops .. by generating all the subarrays !

```python
## Sliding Window Approach Fails for the : Negative Values .. as we have only the condition of while sum > k: reduce 
## Failed for : a = [-1,-1,1] , k = 0 

from typing import List

class Solution:
    def subarraySum(self, a: List[int], k: int) -> int:

        n = len(a)
        count = 0 
        i , j = 0,0
        sum = 0

        if n == 1:
            if a[0] == k:
                return 1
            else:
                return 0
    
        while j < n:
            sum += a[j]

            if sum == k:
                count += 1

            while sum > k:
                sum -= a[i]
                
                if sum == k:
                    count += 1

                i += 1

            j += 1
        
        return  count 
    
```

#### SLIDING WINDOW APPROACH WHICH DOESNT WORKS FOR THE NEGATIVE VALUES : 

```python
if __name__ == '__main__':
    a = [-1,-1,1]
    k = 0
    sol = Solution()
    ans = sol.subarraySum(a, k)
    print(f"Number of ways for the sum {k} is from the array : {a} : {ans}")
```

#### PREFIX SUM SOLUTION : INTUTION

prefixsum : Running sum nikalte jao 

preFixSumOld + k = preFixSumNow

preFixSumOld / NEED = preFixSumNow - k 

Agar hame NEED milta hai.. MAP me matlb.. Jab prefixSumNow ka value NEED hua hoga.. past me jis bhi index pr.. wo index se ab tk .. current ith index tak ke bich ka sum K hai.. and that we can count!


WHY this logic covers ALL subarrays ==> Because every subarray sum can be expressed by subtracting two prefix sums.

```python
class Solution:
    def subarraySum(self, a: List[int], k: int) -> int:

        n = len(a)
        prefixSumNow = 0 
        prefixSumMap = {0: 1}  # sum 0 has 1 occurance : {preSum : 1} : {0: 1}
        
        count = 0 

        for i in a:
            prefixSumNow += i 
            need = prefixSumNow - k 
            if need in prefixSumMap:
                count += prefixSumMap.get(need)
            
            prefixSumMap[prefixSumNow] = prefixSumMap.get(prefixSumNow , 0) + 1   # stores all the prefix sum .. and their count occurance 
        
        return count

if __name__ == '__main__':
    a = [-1,-1,1]
    k = 0
    sol = Solution()
    ans = sol.subarraySum(a, k)
    print(f"Number of ways for the sum {k} is from the array : {a} : {ans}")
```

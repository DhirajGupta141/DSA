##### Search in the 2D Matrix : LEETCODE : 74

```python
from typing import List

class Solution:
    def searchMatrix(self, a: List[List[int]], target: int) -> bool:

        n = len(a)
        m = len(a[0])

        low = 0
        high = n - 1

        row = -1 

        while low <= high:

            mid = low + (high- low)//2

            if (a[mid][0] <= target <= a[mid][m-1]):
                row = mid 
                break
            elif a[mid][0] < target:
                low = mid + 1  
            else:
                high = mid - 1 
         
            
        # proceesing the mid row 

        if row != -1:
            
            low = 0 
            high = m - 1

            while low <= high:
                mid = low + (high- low)//2
                if a[row][mid] == target:
                    return True
                elif a[row][mid] < target:
                    low = mid + 1  
                else:
                    high = mid - 1 

            
        return False 


sol = Solution()

matrix = [[1,3,5,7],[10,11,16,20],[23,30,34,60]]
target = 3

sol.searchMatrix(matrix, target=target)

    
```

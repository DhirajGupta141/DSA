### Evaluate Expression 

Exp 1 : 

a =  ['2','1','+' , '3' , '+'] ==> 
answer : 9 

Ex 2 : 

a = ['4','13',''5,'/','+'] ==> 
answer : 6 

```python
import operator

stack = []
arr = ['2','1','+' , '3' , '*']
# arr = ['4','13','5','/','+']


op = {
    '+' : operator.add,
    '-' : operator.sub,
    '*' : operator.mul, 
    '/' : operator.floordiv  # 3// 2 : 1 , operator.truediv(a,b) ==> 3/2 : 1.5
}

for i in arr: 
    if i.isdigit():
        stack.append(int(i))
    else:
        b = stack.pop()
        a = stack.pop()

        c = op[i](a,b)
        stack.append(c)

print(f"Answer for the expression : {arr} is {stack[0]}")
```

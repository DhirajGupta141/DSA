##### 1047. Remove All Adjacent Duplicates In String

### BRUTE FORCE 

TC : O(n^2)

SC : O(n)

```python

class Solution:
    def removeDuplicates(self, s: str) -> str:

        if len(s) <= 1:
            return s 

        change = True 

        while change:
            change = False
            ans = ''
            i = 0
            
            while i < len(s):
                if i + 1 < len(s) and s[i] == s[i+1]:
                    i += 2
                    change = True 
                else:
                    ans += s[i]
                    i += 1
            
            s = ans 

        return s 

s = 'abbaca'
sol = Solution()
sol.removeDuplicates(s)
```

##### STACK APPROACH 

SC : O(1)

TC : O(1)

```python
class Solution:
    def removeDuplicates(self, s: str) -> str:
        n = len(s)

        if n <= 1:
            return s

        stack = []

        for ch in s:
            if stack and stack[-1] == ch:
                stack.pop()
            else:
                stack.append(ch)

        return ''.join(stack)
    

s = 'abbaca'
sol = Solution()
sol.removeDuplicates(s)  
```

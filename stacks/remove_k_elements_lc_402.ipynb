#### 402. Remove K Digits

Example 1:

Input: num = "1432219", k = 3
Output: "1219"
Explanation: Remove the three digits 4, 3, and 2 to form the new number 1219 which is the smallest.
Example 2:

Input: num = "10200", k = 1
Output: "200"
Explanation: Remove the leading 1 and the number is 200. Note that the output must not contain leading zeroes.
Example 3:

Input: num = "10", k = 2
Output: "0"
Explanation: Remove all the digits from the number and it is left with nothing which is 0.

```python
class Solution:
    def removeKdigits(self, a: str, k: int) -> str:
        n = len(a)
        stack = []

        for i in a:
            while stack and k >=1 and stack[-1] > i:  # here we need while LOOP not if .. we need to keep on removing the elements .
                k -= 1
                stack.pop()
            
            stack.append(i)
        
        # if still k >= 1 : We need to remove the elements from the end as we might have smaller element at the beginning
        while k >= 1:
            stack.pop()
            k -= 1
        
        # result might be empty 
        result = ''.join(stack).lstrip('0')
        
        return  result if result  else '0' 

```

```python
if __name__ == '__main__':
    sol = Solution()
    a  = "1432219"
    k = 3
    ans = sol.removeKdigits(a, k)

    print(f"Smallest Possible value after removing the k digits from {a} is {ans}")
```

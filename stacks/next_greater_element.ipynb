Next Greater Element in an Array !

1. List []  : Act as a Stack --> a[-1] ==> PEEK , a.pop() ==> Deletes the last element of the Array , a.append() ==> Add element at the end
2. Make a Array Circular by Running it twice 

```python
class Solution:

    def next_greater_element_BF(self, a):
        try: 
            print(f"Running BF for the Next Greater Element ")
            n = len(a)
            ans = [-1]*n

            for i in range(0, n): 
                for j in range(i+1,n):
                    if a[j] > a[i]:
                        ans[i] = a[j]
                        break 

            return ans 



        except Exception as e: 
            print(f"Error is in the BF Approach : {e}")
    

    def next_greater_element_stack(self , a):
        """ 
        1. TC : O(n) SC : O(n)
        """
        try: 
           print(f"Running Optimized Approach")

           stack = []
           n = len(a)
           ans = [-1]*n

           for i in range(n-1, -1,-1):
               
               while stack and stack[-1] <= a[i]:
                   stack.pop()
               
               if stack:
                   ans[i] = stack[-1]
               
               stack.append(a[i])

           return ans
        
        except Exception as e: 
            print(f"Error is  Optimized Approach: {e}")
    

    def next_greater_element_stack_circular(self , a):
        """ 
        1. TC : O(2n) SC : O(n)
        """
        try: 
           print(f"Running Optimized Circular Approach")

           stack = []
           n = len(a)
           ans = [-1]*n

           for i in range(2*n-1, -1,-1):
               idx = i % n 
               while stack and stack[-1] <= a[idx]:
                   stack.pop()
               
               if stack:
                   ans[idx] = stack[-1]
               
               stack.append(a[idx])

           return ans
        
        except Exception as e: 
            print(f"Error is  Optimized Approach: {e}")

            
```

```python
if __name__ == '__main__':
    # a = [3,6,1,5,8,4] # [3, 6, 1, 5, 8, 4] Answer :  [6, 8, 5, 8, -1, -1] # Normal 
    a = [3,6,1,5,8,4] # Circular Answer : [6, 8, 5, 8, -1, 6]

    sol = Solution()

    # flag = 1
    flag = 'circular'

    if flag == 0 :
        answer = sol.next_greater_element_BF(a)
    elif flag == 'circular':
        answer = sol.next_greater_element_stack_circular(a)
    else:
        answer = sol.next_greater_element_stack(a)


    print(f" Next Greater elements for the array is {a} is {answer}")
```

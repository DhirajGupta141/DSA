#### Meeting Rooms II (LeetCode 253)

Q. Find How Many rooms required for all the meetings 

Input: [[0,30],[5,10],[15,20]]

Output: 2 ( Meeting Rooms are required to accomodate the above schedule )

Approach Notes : 

Generic : 

0. Below two rules would be true when we have sorted the meetings in the ascending order of the Start time of the meeting rooms .
1. agar koi meeting in-progress hai .. toh next meeting room tab chahiye hoga .. jab pahle wale meeting room ka end_time > start_time of the second meeting 
2. Same old wala meeting room ham tb use kr skte hai fir se .. jab first wale meeting room ka end time < start time of the second meeting room. 

A. MIN HEAP : TC : O(nlogn) --> Sorting and SC: O(n) --> Min heap least used

1. Min heap help isliye kr raha Q ki .. hamne already sort kr rakhha h based on the start time.. So.. Next meeting ko same room me schedule krne ke liye hame dekhna hoga ki pahle ka koi meeting end ho gya hai kya?? That is End time < start time .. of the second.. if yes then we can use the same meeting room. So min heap.. hame Jo meeting end hone wala rahega .. wo deta rhega.. in O(1)
2. So At the end ham log minHeap ka length return kr skte hai as an answer... Q ki.. Jitne Elements ka end time. HEAP me hai.. Utna meetings abhi bhi chal raha hai!
    



```python
import heapq

class Solution:

    def minHeapApproach(self, a):
        try:
            if not a:
                return 0
            
            # sort based on the start time of the Meeting Rooms 
            a.sort(key = lambda x : x[0])

            minheap = []
            heapq.heappush(minheap, a[0][1])
            
            for i in range(1,len(a)):
                
                start , end = a[i]

                # means previous meeting has already ended we can use the same meeting room 
                if start >= minheap[0]:
                    heapq.heappop(minheap)

                heapq.heappush(minheap, end) # added the endtime of the current meet 


            return len(minheap) 
                

        except Exception as e:
            print(f"Error in the optimized approch")
            raise e
    
```

```python
if __name__ == '__main__':
    obj = Solution()
    a = [[0,30],[5,10],[15,20]]
    ans = obj.minHeapApproach(a)
    print(f"For the Schedule {a} : {ans} meeting rooms are required")
```

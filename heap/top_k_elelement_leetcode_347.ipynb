### Leetcode 347 

Heap + map question 

Heap : Storing the frequency and the element in the heap !

Inside the heap.. it does the comparision based on the first element of the tupple if its same then second element of the tuple.

(v1, k1) < (v2, k2) --> Like this comparision happens 

```python
import heapq
from typing import List

class Solution:
    def topKFrequent(self, a: List[int], top: int) -> List[int]:
        
        map = dict()

        for i in a: 
            map[i] = map.get(i, 0) + 1

        minheap = []
        for k , v in map.items():
            heapq.heappush(minheap, (v,k))  # in the heap we are storing the frequency and the element as a tuple (v,k)

            if len(minheap) > top:
                heapq.heappop(minheap)

        return [x[1] for x in minheap] 


"""
TC : O(n) + O(mlogk) --> O(nlogk)
m = unique elements jo heap me jaate hai .. 
logk = heapify of at max k elements 
O(n) = map create 
"""


nums = [1,1,1,2,2,3]
k = 2
print(Solution().topKFrequent(nums, k)) 
```

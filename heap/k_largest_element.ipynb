### Largest and Smallest Element in an Array 

1. For largest Element : Use the min heap 

2. For the smallest Element : use the max heap 

--> Built in Min Heap is there in python 

--> In the python we dont have have the maxheap implemented built-in .. so we use the minheap with a negative value .

--> import heapq


Operation	Complexity
-----------------------

heappush --> O(log n)

heappop	--> O(log n)

heapify	--> O(n)

Access root	O(1)

### Find K smallest elements from an array!

We need to use the Max Heap!

==> TC : O(nlogk) and SC = O(logk)

```python
a = [4,2,5,8,3,1] 
k = 3 # 3 smallest element --> (1,2,3)

import heapq

max_heap = []
for i in a: 
    heapq.heappush(max_heap , -i)  # insert with a negative value !

    if len(max_heap) > k:
        heapq.heappop(max_heap)
    

# now the heapq has 3 smallest value of the array !

ans = [-x for x in max_heap]
ans = sorted(ans) 

print(f"K smallest element of the array {a} is {ans}")
```

### K largest element in an array 

TC : O(nlogk) and SC : O(logk)

```python
import heapq

a = [4,2,5,8,3,1] 
k = 3 # 3 largest element --> (4,5,8)

min_heap = []

for i in a:
    heapq.heappush(min_heap, i)

    if len(min_heap) > k:
        heapq.heappop(min_heap)
    

ans = [x for x in min_heap]
ans = sorted(ans) 

print(f"K Largest element of the array {a} is {ans}")

```

### 3. Compress A String Using Run-Length Encoding (RLE)

Input:  "aaabbcaaa"

Output: "a3b2c1a3"

Notes: 

==> We need to Jump the length of the iteration thats why we need the While Loop in this .


```python
s = "aaabbcaaa"
ans = ""
n = len(s)

i = 0 

while i < n: 
    if i + 1 < n:
        count = 1 
        j = i + 1
        while j < n and s[i] == s[j]:
            j +=1 
            count += 1
        
        ans += s[i] + str(count)
        i = j  # where the j loop fails 

print(f"RLE for the string {s} : is ==>  {ans}")


# TC : O(n * max(Subring having same character)) SC = O(n)
```

### 4. Expand RLE-Compressed String 

Input:  "a3b2c1a3"


Output: "aaabbcaaa"


```python
s = "a3b2c1a3"

n = len(s)
i = 0 
ans = ""

while i < n: 
    if i + 1 < n:
        ans = ans + s[i]*int(s[i+1])
        i += 2

print(f"Expanded RLE for the string {s} : is ==>  {ans}")

```

###  5. Validate Parentheses / Brackets

Input: "{[()]}"

Output: True


```python
mapping = {'{': '}', '(' : ')' , '[' : ']'}

s = '{[()]}'


def validate_paranthesis(s):
    stack = []

    for ch in s:
        print(f"Current char is : {ch}") 
        
        # If it is opening bracket
        if ch in mapping:
            stack.append(ch)

        # If it is closing bracket
        elif ch in mapping.values():
            if not stack:
                return False
            
            last_open = stack.pop()
            
            # Check if correct closing bracket
            if mapping[last_open] != ch:
                return False
        
        else:
            return False   # invalid character

    return len(stack) == 0

answer = validate_paranthesis(s)

if answer:
    print("\n It's a valid Paranthesis")
else:
    print(f"\n Invalid Paranthesis")
    
    


```

7. Reverse a string without affecting special characters

Input:   str = "Ab,c,de!$"

Output:  str = "ed,c,bA!$"

I was Asked in the Media.net Round 1 Interview 



```python
# brute Force 
def reverse_str(s):
    
    ans = ""
    n = len(s)
    special_char_position = []
        
    for i in range(n-1, -1, -1): 
        ch = s[i]
        if ch.isalpha(): 
            ans += ch
        else:
            special_char_position.append(i)
    print(special_char_position)
    
    special_char_position = sorted(special_char_position)
    
    print(f"Special Char position: {special_char_position}")
    
    for i in special_char_position:
        ans = ans[0:i] + s[i] + ans[i:]
        
        
    return ans 
    

s = 'Ab,c,de!$'
ans = reverse_str(s)

print(f"reverese String for input : {s} is {ans}")
```

```python
# reverse using the Single Loop and two Pointer  
# Input:   str = "Ab,c,de!$"
# Output:  str = "ed,c,bA!$"

def reverse_two_pointer(s):
    try:
        print(f"String is  : {s}")

        l = list(s)
        print(f"String List is {l}")

        n = len(l)

        low = 0 
        high = n - 1 

        while low < high:
            if not l[low].isalpha():
                low +=1
            elif not l[high].isalpha():
                high -= 1
            else:
                l[low] , l[high] = l[high] , l[low]   # when both are Alpha-numeric 
                low += 1
                high -=1 

        return "".join(l)
    
    except Exception as e:
        print(f"Error : {e}")
        raise e


s = 'Ab,c,de!$'
ans = reverse_two_pointer(s)

print(f"reverese String for input : {s} is {ans}")
```

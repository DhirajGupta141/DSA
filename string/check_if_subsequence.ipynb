## Check if given String is Subsquence of Other 

Input: s1 = "AXY", s2 = "ADXCPY"

Output: true 

Explanation: All characters of s1 are in s2 in the same order

Input: s1 = "AXY", s2 = "YADXCP"

Output: false 

Explanation: All characters are present, but order is not same.

Input: s1 = "gksrek", s2 = "geeksforgeeks"

Output: true

```python
def isSubsequence(s1,s2):
    try:
        n = len(s1)
        m = len(s2)
        
        if n > m:
            return False

        i = 0  # on s1 
        j = 0  # on S2 

        while i < n and j < m:
            if s1[i] == s2[j]:
                i += 1
                j += 1
            else:
                j += 1
        
        return i == n   # if it would be a subsequence then i would have compeletely iterated through String S1
          
    except Exception as e:
        print(f"Error : {e}")
        raise e
    


s1 = 'abc'
s2 = 'nadbfgce'

isSubsequence(s1,s2)
```

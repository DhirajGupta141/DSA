##### Group Anagram LC : 49

```python
class Solution:
    def groupAnagramBF(self, strs):
        try: 
            n = len(strs)

            curr_dict = dict()
            for s in strs:
                ss = ''.join(sorted(s)) # sorted returns a list of characters so we need to join them 
                if ss in curr_dict:
                    curr_dict.get(ss).append(s)
                else:
                    curr_dict[ss] = [s]

            ans = []
            for key , value in curr_dict.items():
                ans.append(value)
            
            return ans
        except Exception as e:
            raise e 
    
    def groupAnagramOp(self , strs):
        try: 
            n = len(strs)

            curr_dict = dict()

            for s in strs:
                a = [0]*26
                for ch in s:
                    a[ord(ch) - ord('a')] += 1
                
                a = tuple(a)  # list cannot be used as key for the dict as list are mutable 

                if a in curr_dict:
                    curr_dict[a].append(s)
                else:
                    curr_dict[a] = [s] 
            
            return list(curr_dict.values())
        except Exception as e:
            raise e 
        
    
```

#### String to Integer LC : 8 

This is AWESOME solution of the Iterative Approach Solved this buy myself after 9 Submission .. Then it cleared all 1200 TC 

```python
class Solution:
    def myAtoi(self, s: str) -> int:

        s = s.strip() # remove the trailing and leading spaces 
        
        negative = False 
        n = len(s)

        ans = ""

        for i , c in enumerate(s):
            if i == 0 and c == '-':
                negative = True 
            elif c.isalpha() or c in ['.','-', ' '] or (c == '+' and i > 0):
                break
            else:
                if ans == "" and (c == '0' or c == '+'): 
                    continue
                ans += c
        
        if not ans:
            return 0 
        
        if negative:
            ans = -int(ans)
            if ans < -2**31:
                return -2**31
            else:
                return ans         
        
        return 2**31-1 if int(ans) >= 2**31 else int(ans)


s = '    23bc'
sol = Solution()
ans =sol.myAtoi(s)
ans 

```

### Check if the String is palindrome by removing at most one character 


input : "abceba"

Output : True

### Brute Force Solution 

1. removing each element one by one .. and checking if we can make the string Palindrome .. by that deletion 

```python
def is_palindrome(s):
    return s == s[::-1]

def canWe(s):

    # if s iteself is palindrome 
    if is_palindrome(s):
        return True 
    
    n = len(s)
    for i in range(0,n-1):
        new_s = s[0:i] + s[i+1:]   # remove of the ith character 
        if is_palindrome(new_s):
            return True 
    
    return False
            
        
s = "abcedba"        
canWe(s)
```

### The Below one is the Optimized Solution !

```python
def is_palindrome(s, i , j):

    while i < j and s[i] == s[j]:
        i += 1
        j -= 1
    
    return i >= j


def canThisBePalinedrome(s, i , j):
    while i < j:
        if s[i] != s[j]:
            return is_palindrome(s, i+1 , j) or is_palindrome(s , i, j-1)   # we need to validate the remaining string .. after one element delete either left or the right character 
        i += 1
        j -= 1
    
    return True

```

```python
if __name__ == '__main__':
    s = "abba"
    n = len(s)

    # answer = is_palindrome(s , 0 , n-1)
    answer = canThisBePalinedrome(s,0,n-1)

    if answer:
        print(f"Given String can be made Palindrome")
    else:
        print(f"Given String cannot be made Palindrome")
```

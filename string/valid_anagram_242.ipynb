##### Valid Anagram : LC : 242

```python


class Solution:

    # Using the sort function : 
    def isAnagram(self, s: str, t: str) -> bool:
        
        n = len(s)
        m = len(t)

        if n != m:
            return False

        s = sorted(s)   
        t = sorted(t)

        return s == t


    def isAnagramOp(s ,t):
        n = len(s)
        m = len(t)

        if n != m:
            return False 
        
        a = [0]*26

        for c in s:
            a[ord(c) - ord('a')] += 1 
        
        for c in t:
            a[ord(c) - ord('a')] -= 1
        
        for val in a:
            if val != 0:
                return False

        return True 

s ="banana"
t = "ananab"

sol = Solution()
ans = sol.isAnagram(s, t)
ans 

        
```

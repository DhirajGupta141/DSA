##### Longest Palindromic Substring : LC : 5

Identification : 
1. Double for loop can solve this 
2. Palindrome -> 2 Pointers 

```python

"""
BF : By Generating all the substrings .. Using the double for loop and checking if they are panindrome and 
     then finding max of all those which are palindrome 

    TC : O(n^3) --> n^2 for the generate and then n for the compare 
    SC : O(1)
"""
class Solution:
    def longestPalindrome(self, s: str) -> str:

        n = len(s)
        if n <= 1:
            return s

        ans = ""
        for i in range(n):
            for j in range(i, n):
                substring = s[i:j+1] # Generating all the substring staring from i till j 

                if substring == substring[::-1]:  # check if : ulta sidha ek saman 
                    if len(substring) > len(ans):
                        ans = substring
        
        return ans 
                
s = "ababcc"
sol = Solution()
ans = sol.longestPalindrome(s)
print(f"For the String : {s} the longest palindromic substring is ==> {ans}")
        
```

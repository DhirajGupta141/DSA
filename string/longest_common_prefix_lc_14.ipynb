#### Longest Common Prefix : LC : 14 

Input: strs = ["flower","flow","flight"]

Output: "fl"

Constrains : 

1 <= strs.length <= 200

0 <= strs[i].length <= 200

strs[i] consists of only lowercase English letters if it is non-empty.

#### 1. Brute Force Solution using the For Loop 

TC : O(n*m)

m = Longest Prefix Length 

SC : O(1)

#### 2. Sorting Technique : 

TC : O(m*nlogn) m = Longest Common Prefix 

SC : O(1)

```python
class Solution:

    def bruteForce(self, s):
        try: 
            n = len(s)

            if n == 0:
                return ""
            if n == 1:
                return s[0]
            
            ans = float('inf')
            first = s[0]

            for i in range(1, n):

                j = 0 
                second = s[i]

                while j < len(first) and j < len(second) and first[j] == second[j]:
                    j += 1
                
                ans = min(ans , j)   # we need the Longest common in all the elements in the string 
            
            return first[:ans]

        except Exception as e:
            raise e
        
    

    def sortingTechnique(self , s):
        try:
            n = len(s)

            if n == 0:
                return ""
            if n == 1:
                return s[0]

            s.sort() # sort inplace returns nothing unlike a = sorted(a)

            first = s[0]
            last = s[-1]

            loop_len = min(len(first) , len(last))

            ans = 0
            for i in range(loop_len):
                if first[i] != last[i]:
                    return first[:ans]
                ans += 1
            
            return first[:ans]

        except Exception as e:
            raise e 
```

```python
if __name__ == '__main__':
    sol = Solution()

    s = ["flower","flow","flight"]
    # ans = sol.bruteForce(s)
    ans = sol.sortingTechnique(s)
    print(f"Longest Common Prefix for the strings {s} is {ans}")
    
```

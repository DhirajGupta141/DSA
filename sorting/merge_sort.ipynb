### Merge Sort : Divide and then Conquor 

TC : O(nlogn)

SC : o(n) --> Ans List used 

1. keep diving the array till it has only one element(base case).

2. Then merge kr do outputs ko 


```python
class Merge:

    def divide(self, a):
        n = len(a)

        # Base case.. of the division
        if n <= 1:
            return a

        # Hypothesis
        mid = n//2
        left_array =  self.divide(a[:mid])
        right_array = self.divide(a[mid:])

        # conduction / Answer Calc 
        return self.merge(left_array,right_array)


    def merge(self, left , right):
        ans = []
        i = 0
        j = 0 
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                ans.append(left[i])
                i += 1
            else:
                ans.append(right[j])
                j += 1
        
        # Extend the list .. by the remaining Elements in the left and the right array 
        if i < len(left):
            ans += left[i:]

        if j < len(right):
            ans += right[j:]

        return ans 
        



```

```python
if __name__ == "__main__":
    a = [4,3,8,1,7]
    obj = Merge()
    ans = obj.divide(a)
    
    print(f"Sorted array for the input array {a} : is : {ans}")
```

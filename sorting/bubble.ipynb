### Bubble Sort

TC : O(n^2)

Intution : 

1. We start the array from the index = 0 .. we find the place for the ith element.. by comparing with the adjecent element 

2. Har ek iteration me.. elements sort hote jate hai..

3. First iteration me.. largest element apne position pr aayega .. uske baad second larget uske baad 3rd larget.. aaise krte krte sab sort ho jayenge .. isliye ham likhet hai .. range(0 to n-i-1) --> Q ki last i element apne hota h .. har ek ith iteration pr !

```python
a = [4,3,8,2,1,7]
input = [4,3,8,2,1,7]

n = len(a)

for i in range(n):
    for j in range(0,n-i-1):
        if a[j] > a[j+1]:    # For Decending just change this > to <
            a[j] , a[j+1] = a[j+1] , a[j]

print(f"Sorted array for the input array {input} : is : {a}")

```

### Bubble Optimized 

```python
a = [1, 2, 3, 4, 7, 8] #[4,3,8,2,1,7] 
input = [1, 2, 3, 4, 7, 8] 

n = len(a)
swap = False 

for i in range(n):
    for j in range(0,n-i-1):
        if a[j] > a[j+1]:
            a[j] , a[j+1] = a[j+1] , a[j]
            swap = True
    
    if not swap:
        print(f"Array is already sorted ")
        break 



print(f"Sorted array for the input array {input} : is : {a}")

```

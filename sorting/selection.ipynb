### Selection Sort!

TC : O(n^2)
SC : O(1)

1. Har ek iteration me .. ham min element find krte h.. and Min element ko ith element k sath swap kr dete h ..

2. Finally Sare elements sort ho jate hai! 

```python
a = [4,3,8,2,1,7]
input = [4,3,8,2,1,7]

n = len(a)


for i in range(n):
    min_element = i   # finding the position of this element 
    for j in range(i+1 , n):
        if a[j] < a[min_element]:   # just change < to > to Make it .. Decending order 
            min_element = j 
    
    a[i] , a[min_element] = a[min_element] , a[i]


print(f"Sorted array for the input array {input} : is : {a}")


```

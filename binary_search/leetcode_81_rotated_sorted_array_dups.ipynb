##### 81. Search in Rotated Sorted Array II

```python
from typing import List
class Solution:
    def search(self, a: List[int], target: int) -> bool:

        n = len(a)
        left = 0 
        right = n - 1

        while left <= right :
            mid = (left + right )//2 

            if a[mid] == target:
                return True 


            # Skipping the duplicate part of the array : Not able to get the sorted half
            if a[left] == a[mid] == a[right]:
                left += 1
                right -= 1
                
            # case 1 : Left half is sorted : as left <= mid --> elements from the left till the mid are sorted 
            elif a[left] <= a[mid]:
                if a[left] <= target < a[mid]:
                    right = mid - 1
                else:
                    left = mid + 1 
            else:
                # case 2 : elements from the mid to the right are sorted 
                if a[mid]  < target <= a[right]:
                    left = mid +  1
                else:
                    right = mid - 1 
        
        return False 


nums = [2,5,6,0,0,1,2]
target = 2
sol = Solution()
sol.search(nums, target )
```

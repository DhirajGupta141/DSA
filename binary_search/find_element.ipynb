##### Binary Search : making O(n) solution to O(nlogn)

```python
class Solution:

    def binarySearch(self , a, k ):
        try: 
            n = len(a)
            i = 0 
            j = n - 1

            while i <= j:
                mid = i + (j - i )// 2
                if a[mid] == k:
                    return True 
                elif a[mid] < k:
                    i = mid + 1
                else:
                    j = mid - 1
            
            return False 
        except Exception as e:
            raise e
        
a = [1,2,3,4,5,7,10]
k = 2
sol = Solution()
ans = sol.binarySearch(a,k)
ans  
```

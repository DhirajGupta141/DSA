#### Subarray Of Size k with the given SUM 

https://www.geeksforgeeks.org/dsa/subarray-of-size-k-with-given-sum/

```python
class Solution:
    def bf(self, a,k,s):
        """ Generating all the subarray !
        TC : O(n*k)
        SC : O(1)
        """
        try:
            n = len(a)

            for i in range(n-k+1):
                curr_sum = 0 
                for j in range(i,i+k+1):
                    curr_sum += a[j]
                if curr_sum == s:
                    return True

            return False 
                

             
        except Exception as e:
            raise e
    

    def sw(self , a, k , s):
        """ This is SW solution Where we are mainting the Window of Size K By reducing the Processed Elements
        TC : o(n)
        SC : O(1)
        """
        try:

            n = len(a)

            curr_sum = 0
            for i in range(k):
                curr_sum += a[i]

            if curr_sum == s:
                return True 
            
            print(f"Current Sum is {curr_sum}")

            j = k
            while j < n:
                curr_sum = curr_sum + a[j] - a[j-k]
                if curr_sum == s:
                    return j, True 
                j += 1
            
            return False



        except Exception as e:
            raise e
        

a = [1,4,2,10,2,3,1,0,20]
k = 4
s = 18

sol = Solution()

flag = 0

if flag:
    print(f"BF : answer : {sol.bf(a,k,s)}")
else:
    print(f"SW : answer : {sol.sw(a,k,s)}")
```

##### LC : 3 : Longest Substring Without Repeating Characters

#### BF : 

Generating all the substrings 

#### SW : 

Mainting a Window of unique Elements : Removing the elements from the left if there is any character at the right side which is making the window to have dups

```python

class Solution:
    def bf(self, s: str) -> int:

        n = len(s)

        if n <= 1:
            return n 
        
        ans = 0
        for i in range(n):
            curr_dict = dict()
            for j in range(i , n):
                if s[j] in curr_dict:
                    break
                else:
                    curr_dict[s[j]] = 1
            
            ans = max(ans , len(curr_dict))

        
        return ans 

    # Sliding Window  TC : O(2n) -> O(n) and SC : O(n) ==> 2n because of Right pointer operation and left one operation 
    def sw(self, s: str) -> int:

        n = len(s)
        if n <= 1:
            return n 
        
        i = 0 
        j = 0 

        ans = 0
        curr_set = set()  # maintaining the unique elements so far

        while j < n:
            
            # this s[j] is making the set filled with dupes so remove the old one and fill with the new one 
            while s[j] in curr_set:   
                curr_set.remove(s[i])
                i += 1 

            curr_set.add(s[j])

            # this creates the window of Uniue elements 
            ans = max(ans , j - i + 1)  # as Usual Window Size j - i + 1
            j += 1
        
        return ans 

sol = Solution()
flag = 0
s = "abcabcbb"

if flag == 1:
    ans = sol.bf(s)
    print(f"BRUTEFORCE Ans for the string {s} is {ans}")
else:
    ans = sol.sw(s)
    print(f"SLIDING WINDOW : Ans for the string {s} is {ans}")
```

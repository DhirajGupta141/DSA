#### Leetocode : 487 Premium : Maximum Consecutive ones with at most k flip of 0 to 1 

##### Brute Force Solution Using the Loops and by generating all the subarrays of consecutive ones by having the flip of Zeros 

```python
from typing import List

class Solution:
    def findMaxConsecutiveOnesKFlipBF(self, a: List[int], k) -> int:
        try:
            n = len(a)
            ans = 0 

            for i in range(n):
                zeroCount = 0    # At each Subarray Start we have K Flips available 
                for j in range(i+1 ,n):
                    
                    if a[j] == 0 and zeroCount < k:
                        zeroCount += 1
                    
                    ans = max(ans , j - i + 1)
                    
                    if zeroCount >= k:
                        break
            
            return ans 
        except Exception as e:
            raise e



a = [1,1,0,1,0,0,0,1,1]
k = 2
sol = Solution()
ans = sol.findMaxConsecutiveOnesKFlipBF(a, k)
ans     
```

### Why Sliding Winodw (SW) ?

1. Array 
2. k
3. Maximum 
4. All positive 
5. Consecutive / Subarray

```python
from typing import List

class Solution:
    def findMaxConsecutiveOnesKFlip(self, a: List[int], k) -> int:
        try:
            i = 0 
            j = 0 
            n = len(a)
            zeroCount = 0 
            ans = 0 

            while j < n:
                if a[j] == 0:
                    zeroCount += 1
                
                while zeroCount > k:
                    if a[i] == 0:
                        zeroCount -= 1
                    i += 1

                ans = max(ans , j - i + 1)   # why end ? Because above we are making sure there are only <= k zeros are in the window 
                j += 1
            
            return ans 


        except Exception as e:
            raise e
        

a = [1,1,0,1,0,0,0,1,1]
k = 2
sol = Solution()
ans = sol.findMaxConsecutiveOnesKFlip(a, k)
ans
```

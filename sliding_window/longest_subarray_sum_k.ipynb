### Brute Force

1. Generate All the subarray from each of the ith element till the end of the array .. and find the window ..

```python
class LongestSuarraySumK:

    def bruteForce(self,a , k):
        try:
            """ Generating all the subarray using the for loop"""

            n = len(a)
            maxiLen = 0 
            start = 0
            end = 0 

            for i in range(0,n):
                sumi = 0
                for j in range(i,n):  # Generating all the subarray starting for the ith Index 
                    sumi += a[j]
                    if sumi == k:
                        if maxiLen < (j-i-1):
                            start = i 
                            end = j 

            return a[start:end+1] if end > 0 else []
        
        except Exception as e:
            print(f"Error while computing the brute force")
    
    def slidingWindow(self,a , k):
        try:
            n = len(a)
            maxiLen = 0 
            start = 0
            end = 0 

            sumi = 0 
            
            # Standard way of initializing the Sliding Window  Technique
            i = 0 
            j = 0 

            while j < n:
                sumi += a[j]
                
                # we found the Sum 
                if sumi == k:
                    if maxiLen < (j-i-1):
                        start = i 
                        end = j 
                
                # Till Sumi is more than the desired val   = Reducing the winodw 
                while sumi > k:
                    sumi -= a[i]
                    i += 1
                
                j += 1  # expanding the window 
            
            return a[start:end+1] if end > 0 else []
                


        except Exception as e:
            print(f"Error while computing the Optimized Solution ")
            raise e
```

```python
if __name__ == '__main__':
    
    obj = LongestSuarraySumK()

    nums = [10, 5, 2, 7, 1, 9]
    k = 15  
    
    flag = 1 

    if flag == 0:
        ans = obj.bruteForce(nums, k)
        print(f"Brute force solution answer is {ans}")
    else:
        ans = obj.slidingWindow(nums, k)
        print(f"Optimized solution answer is {ans}")

```

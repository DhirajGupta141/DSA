#### 646. Maximum Length of Pair Chain

Example 1:

Input: pairs = [[1,2],[2,3],[3,4]]

Output: 2

Explanation: The longest chain is [1,2] -> [3,4].

Example 2:

Input: pairs = [[1,2],[7,8],[4,5]]

Output: 3

Explanation: The longest chain is [1,2] -> [4,5] -> [7,8].

### How to Decide  start or end which one for the sorting ?

EXTEND ? - END : selecting the pair that ends earliest leaves maximum room to add more pairs

Sorting by Start would fail because .. We need to extend the length of the chain .. because for the 

input.. : [[1,100], [2,3], [4,5], [6,7]]

After sort : [[1,100], [2,3], [4,5], [6,7]]

Now .. To extend.. if we select [1,100] as start of the chain.. we cannot select any other element after this 




```python
from typing import List 

class Solution:
    def findLongestChain(self, a: List[List[int]]) -> int:

        n = len(a)

        a.sort(key = lambda x : x[1])

        chain = [0]
        for i in range(1,n):
            index = chain[-1]
            previous = a[index]
            current = a[i]

            if current[0] > previous[1]:
                chain.append(i)
        
        return len(chain)  , chain
```

```python
if __name__ == '__main__':
    sol = Solution()
    a = [[1,2],[2,3],[3,4]]
    ans, chain = sol.findLongestChain(a)
    print(f"Longest Chain that can be formed Using the array {a} is having length {ans} and the index of these elements are {chain}")
```

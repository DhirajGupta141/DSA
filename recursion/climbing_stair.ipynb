### Clibming Stair with Maximum Steps 

1. ways(h) = ways(h-1) + ways(h-2) + ... + ways(h-maxSteps)
2. dp[i] = dp[i-1] + dp[i-2] + ... + dp[i-maxSteps]


Why the above logic works ?

1. Assume k = 2 : Agar lets say hame 3rd step pr aana hai.. then ham 1st and 2nd steps pr se hi aa skte h .. 3rd pr right because K = 2
2. But Fir 1st and 2nd setps .. par aane ka answer toh hame already calculate kr liya hai (Overallping Subproblem) .. So why not to use that.
3. 100 steps pr jaane ke liye ham.. 98 and 99 setsp se hi jaa skate h.. so.. mai fir se 0th step pr.. se jaa kr Q calc kru ki 100 pr jaane k liye kitna ways h... 100th = 98th pr jitna hai + 99th pr jitna hai 

```python
class Solution: 

    def recursive_sol(self , n, k): 
        try: 

            print(f"Recur sol with k = 2")
            
            if n <= 1: 
                return 1

            # because I know k would be <= 2        
            answer = self.recursive_sol(n-1, k) +  self.recursive_sol(n-2, k) 

            return answer 
        
             
        except Exception as e: 
            print(f"error is {e}")

        
    def recursive_sol_2(self , n, k): 
        """
        1. Agar K ka Value Bahhut jayada ho like 5 . Then 5 baar call nhi krenge function right? Islite loop me. Run 
        2. This is a Solution without the memoization
        3. Har ek Recursive call me .. ye function ka Ek copy banta h... and har ek copy .. or child apne parent ko apana ways return krta hai  
        4. Jo skba parent ha wo hai.. Jo hamne Main() function se call kiya h.. that is recursive_sol_2(4,2) --> And ye apne childs ka answer laa kr return kr dega final result.
        """
        try: 

            print(f"Recur Sol Without Memo")


            if n <= 1: 
                return 1
            
            
            ways = 0 
            for step in range(1, k + 1):     
                ways += self.recursive_sol_2(n - step, k)

            return ways 
        
            
        except Exception as e: 
            print(f"error is {e}")   

    def recursive_sol_memo(self , n, k,memo): 

        """
        1. Agar K ka Value Bahhut jayada ho like 5 . Then 5 baar call nhi krenge function right? Islite loop me. Run 
        2. This is a Solution with Memoization 
        """
        try: 

            print(f"Recur Sol With memo")

            if n <= 1: 
                memo[n] = 1
                return 1
            
            if n in memo:
                return memo[n]
            
            ways = 0 
            for step in range(1, k + 1):     
                ways += self.recursive_sol_memo(n - step, k, memo)
            
            memo[n] = ways 

            return ways 
        
            
        except Exception as e: 
            print(f"error is {e}")   
        
    
    def array_sol(self , n, k ): 
        try: 
            
            print(f"Dyanmic Progamming using the Array!")
            a = [0]*n
            a[0] = 1
            a[1] = 2 

            if n == 1: 
                return 1 
            elif n == 2: 
                return 2
            
            print(f"Array value is : {a}")
            
            for i in range(2, n):

                ways = a[i-1] + a[i-2]
                a[i] = ways 
                print(f"Steps req. for the ith : {i+1} step  is {ways}")

            print(f"Array Value after Iteration is : {a}")

            return a[n-1]   
        
        except Exception as e: 
            print(f"error is {e}")
```

```python
if __name__ == '__main__':
    n = 4 # 4 : 5, 7 : 21,9 : 55
    k = 2 

    memo = {}

    # flag = 'recursion'
    # flag = 'recursion_2'
    flag = 'recursion_3'

    sol = Solution()

    if flag == 'recursion':
        answer = sol.recursive_sol(n,k)
    elif flag == 'recursion_2':
        answer = sol.recursive_sol_2(n, k)
    elif flag == 'recursion_3':
        answer = sol.recursive_sol_memo(n,k, memo)
    else:
        answer = sol.array_sol(n,k)

    print(f"Number of Ways : {answer} ")
```

#### Rotate Array By K Position Towards Right : Leetcode : 189

a : [1,4,10,0,30,5]

k = 2 

Output : [30,5,1,4,10,0]

```python
## This solution is of Using the Extra Space of O(n) Which Is not ab Optimized Solution 
def rotate_array(a,k):
    try:
        n = len(a)

        k =  k%n  # case when k > n we dont want that circular operation

        ans = a[-k:]  + a[:n-k]   # concat of the last k elements and the inital n-k elements 

        print(f"NEW ARRAY IS : {ans}")

        return ans 
    except Exception as e:
        print(f"Error is {e}")
        raise e

a = [1,4,10,0,30,5]
k = 2 
new_array  = rotate_array(a , k)

print(f"Rotated Array for the Input Array : {a}  by {k} position is {new_array}")
```

### Optimized Approach : In-place 

```python
def reverse(a, i, j):
    while i <=  j:
        a[i] , a[j] = a[j], a[i]
        i += 1
        j -= 1
    

def rotate_array_optimized(a, k):
    try: 
        n = len(a)
        k = k % n
        
        # Using Built in REVERSE FUNCTION 
        """
        a.reverse()
        a[0:k] = a[0:k][::-1]
        a[k:] = a[k:][::-1]
        """
        
        # USING SELF REVRESE FUNCTION 
        reverse(a,0,n-1)  # Reverse Entire array 
        reverse(a,0,k-1) # First K elements 
        reverse(a,k,n-1) # last n-k elements 

        print(f"Rorated Array InPlace is : {a}")
    except Exception as e:
        print(f"Error is {e}")
        raise e


a = [1,4,10,0,30,5] # [30, 5, 1, 4, 10, 0]
k = 2 
rotate_array_optimized(a , k)
    
```

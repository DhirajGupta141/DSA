### Common Key Pairing 

```python
# input 
list1 = [(1, "a"), (2, "b"), (4, "c")]
list2 = [(2, "e"), (4, "g"), (7, "h")]

# Output : [(2, ('b', 'e')), (4, ('c', 'g'))]
```

```python
list1
```

```python
list2 
```

```python
dict1 = dict(list1)
dict1
```

```python
dict2 = dict(list2)
dict2
```

```python
ans = []
for k in dict1:
    if k in dict2:
        ans.append((k , (dict1[k], dict2[k])))

ans 
```

```python
# one liner 
[(k , (dict1[k], dict2[k])) for k in dict1.keys() & dict2.keys()]

```

### Without Dict 

```python
list1 = [(1, "a"), (2, "b"), (4, "c")]
list2 = [(2, "e"), (4, "g"), (7, "h")]
```

```python
ans = []
for k1 , v1 in list1:
    for k2 , v2 in list2:
        if k1 == k2:
            ans.append((k1,(v1,v2)))

ans
```

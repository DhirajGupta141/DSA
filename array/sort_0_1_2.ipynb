#### SORT 0 ,1 2 using 3 Pointers 

```python
from typing import List

class Solution:
    def sortColors(self, a: List[int]) -> None:
        
        n = len(a)

        i = 0  # manage 0 
        j = 0  # manage 1
        k = n - 1 # manage 2 

        while j <= k: 

            if a[j] == 0:  
                a[i] , a[j] = a[j] , a[i]
                i += 1
                j += 1   # after swap inc. j because we already know what was there at th ith position as we have already procesed it 

            elif a[j] == 2:
                a[k] , a[j] = a[j] , a[k]
                k -= 1  # We are swapping with k so we dont know what k brings so j cannot be inc.
            else:
                j += 1

a = [2,1,0,2,1,0,2]
sol = Solution()
sol.sortColors(a)

print(f"Sorted array is : {a}")
```

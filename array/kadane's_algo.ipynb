# Maximum Subarray Sum problem  : Leetcode 53 

Brute Force : 

1. Generating all possible subarray using the i and j pointers --> TC = O(n^2)
2. Using the Kadane's Algo.. TC = O(N)

#### Kadane's algo : 

1. At each element in the array we need to make a choice.. existing subarray ke sath continue krna hai ya fir current element k sath ek naya array banana start krna hai ?

2. Since hame answer ko maximize krna hai .. so ham hamesha max lenge .. and sum = max(a[i], a[i] + sumi) and 
then maxi = max(maxi , sumi)



```python
class Solution:

    def maximumSubarraySumBF(self, a):
        try: 
            n = len(a)

            maxii = float('-inf')

            for i in range(0 , n): 
                sum = 0 
                for j in range(i, n):
                    sum += a[j]
                
                    maxii = max(maxii , sum )
            
            return maxii

        except Exception as e: 
            print(f"Error is : {e}")
    

    def maximumSubarraySumOptimized(self , a):
        try: 
            # Kadane's Algorithm 
            n = len(a)
            maxi = float('-inf')  # as we can have all the elements in the array as negative 
            sum = 0 

            for i in range(0, n): 
                sum += a[i]

                maxi = max(maxi , sum )

                # Now we would look for another subarray 
                if sum < 0: 
                    sum = 0 

            return maxi 
        
        except Exception as e: 
            print(f"Error is : {e}")
    
    # Kadane's 
    def kadane(self, a):
        sumi = a[0]
        maxi = a[0]

        for i in range(1,len(a)):
            sumi = max(a[i], a[i]+ sumi) # making a choice to continue the current subarray or start a new subarray from the current ele
            maxi = max(maxi , sumi)
        
        return maxi 
```

```python
if __name__ == '__main__':
    a = [2,3,-7,4,7,-1]

    sol = Solution()

    flag = 1

    if flag == 0 :
        print(f"Ran the bruteforce Approach for this probelm!")
        answer = sol.maximumSubarraySumBF(a)
    else:
        print(f"Ran the Optimized Approach for this probelm!")
        # answer = sol.maximumSubarraySumOptimized(a)
        answer = sol.kadane(a)


    print(f"Maximum Subarray Sum for the array is {a} is {answer}")
```

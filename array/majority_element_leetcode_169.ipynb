### Majority Element 169 

Given an array nums of size n, return the majority element.

The majority element is the element that appears more than ⌊n / 2⌋ times. You may assume that the majority element always exists in the array.

 

Example 1:

Input: nums = [3,2,3]
Output: 3

Example 2:

Input: nums = [2,2,1,1,1,2,2]
Output: 2


## Moore's Alogorithm 

"""
Why This Works : 

1. n/2 se jayada count hone ke liye.. har ek majority element dusre elements ko cut krte jayega .. and last me wahi bachega. [2,2,1,1,1,2,2] Jaise yaha 2 ne 1 ko cut kiya but last me ek bach gya!
"""

















        

```python

class Solution:
    def majorityElement(self, a: list[int]) -> int:

        n = len(a)
        if n <=1:
            return a[0]

        count = 1 
        ele = a[0]

        for i in range(1,n):
            if a[i] == ele:
                count += 1
            else: 
                count -= 1
                if count < 0: # means the current element would be majority before this everything is cut 
                    count = 1
                    ele = a[i]
        
        return ele


```

```python
if __name__ == '__main__':
    a = [2,2,1,1,1,2,2]
    obj = Solution()

    ans = obj.majorityElement(a)

    print(f"Majoirty element for the array : {a} is {ans}")
```

##### LC : 918 : Maximum Sum Circular Subarray

##### Brute Force Solution - 

Circular using idx = (i+j)% n 

```python
from typing import List 

class Solution:
    def maxSubarraySumCircular(self, a: List[int]) -> int:
        
        n = len(a)
        ans = a[0]

        if n == 1:
            return a[0] 

        
        for i in range(n):
            sumi = 0 
            for j in range(n):   # always runs from 0 
                idx = (i+j)%n    # Making it circular 
                sumi+= a[idx]

                ans = max(ans , sumi)

        return ans 
    
a = [2,2-3,4,5]
sol = Solution()
ans = sol.maxSubarraySumCircular(a=a)
print(f"Maximum Circular Subarray sum for the array : {a} is {ans}")
```

##### Circular Maximum Subarray Sum Using the kadane's Algorithm 

```python

class Solution:
    def maxSubarraySumCircular(self, a: List[int]) -> int:
        
        n = len(a)

        if n ==1 :
            return a[0] 

        
        maxSubarray = a[0]
        minSubarray = a[0]
        maxSum = a[0]
        minSum = a[0]

        totalSum = a[0]

        for i in range(1,n):

            # maxisum Subarray Sum 
            maxSum = max(maxSum + a[i], a[i])
            maxSubarray = max(maxSubarray , maxSum)

            # minum Subbary Sum 
            minSum = min(minSum + a[i], a[i])
            minSubarray = min(minSubarray,minSum)

            # total Array Sum 
            totalSum += a[i]

        # Out of the total We are skipping some portion of the array to make it max sum ( Some from left , some from right , subtract sum mid )
        circular_max_sum = totalSum - minSubarray  # -14 - (-14) = 0 

        if circular_max_sum == 0:   # when all the array elements are negative 
            return maxSubarray 
        
        
        return max(circular_max_sum , maxSubarray)

a = [2,2-3,4,5]
sol = Solution()
ans = sol.maxSubarraySumCircular(a)

print(f"Maximum Circular sum for the array : {a} is {ans}")
```

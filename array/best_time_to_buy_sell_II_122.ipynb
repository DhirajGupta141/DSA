## 122 : Best time to Buy and Sell the Stocks : Multiple Buy Sell Allowed

when we buy a stock profit goes in negative else postive when we sell the stock

holding = 1 --> we have bought previously so we can sell now  
holding = 0 --> we have not bought previously so we need to buy first , and then we can sell


At each Index we can either : Skip or sell | skip or Buy --> as the buy and sell toggles 

BruteForce : 

TC : O(2^n) --> Each index we have 2 choice ..
 
SC : O(n) --> Auxilary recursion stack 


##### BruteForce 

```python
from typing import List

class Solution:
    def maxProfit(self, a: List[int]) -> int:

        def dfs(i ,holding):

            if i == len(a):
                return 0

            skipAns = dfs(i+1 , holding)

            if holding:
                # sell 
                sellAns = a[i] + dfs(i+1 , 0)

                return max(skipAns , sellAns)

            else:
                # buy 
                buyAns = -a[i] + dfs(i+1 ,1)
                return max(skipAns , buyAns)

        return dfs(0,0) 
    
a = [7,1,5,4,3,6]
sol = Solution()
ans = sol.maxProfit(a)
ans 
```

#### Memoization 


TC : O(n) * 2 --> 2n --> O(n)

SC : O(n) * 2 --> O(n)

2 because i can go from 0 to n and at each of the index we have 2 choice either to buy or selll


```python
from typing import List

class Solution:

    def maxProfit(self, a: List[int]) -> int:
        cache = dict()
        def dfs(i, holding):
            if i == len(a):
                return 0

            if (i,holding) in cache:
                return cache[(i,holding)]

            skipAns = dfs(i+1 , holding)

            if holding:
                sellAns = a[i] + dfs(i+1 , 0) # toggle 
                ans = max(skipAns , sellAns)
                
            else:
                buyAns = -a[i] + dfs(i+1 , 1) # toggle
                ans = max(skipAns , buyAns)

            # Store the result before returing the answer for the ith position 
            cache[(i, holding)] = ans 
            return ans 

        return dfs(0,0)

a = [7,1,5,4,3,6]
sol = Solution()
ans = sol.maxProfit(a)
ans 
```

#### Greedy : Logical , But not Practical 


1. ye bas aaise hai ki .. small small wins ko sum up kr rahe .. rather than finding the optimal buy sell at once 
2. Ye wala logical solution hai na ki practicel .. [1,2,3,4,5]

Greedy : 

2-1 : 1

3-2 : 1

4-3 : 1

5-4 : 1

Total : 4

other way :

5-1 = 4 


```python

from typing import List
class Solution:
    def maxProfit(self, a: List[int]) -> int:

        ans = 0 
        n = len(a)

        for i in range(1,n):
            if a[i] > a[i-1]:
                ans = ans + (a[i]-a[i-1])

        return ans  


a = [7,1,5,4,3,6]
sol = Solution()
ans = sol.maxProfit(a)
ans 

```

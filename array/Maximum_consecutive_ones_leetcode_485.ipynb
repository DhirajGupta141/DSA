##### Leetcode : 487 : Maximum Consecutive Ones 


1. Logic of Continuing the Subarray if condition met else stop it 

```python
from typing import List

class Solution:
    def findMaxConsecutiveOnes(self, a: List[int]) -> int:

        n = len(a) 
        ans = 0 
        count = 0
        for i in a:
            if i == 1:
                # continue the subarray of consecutive one 
                count += 1
                ans = max(ans , count)
            else:
                # consecuttive one's subarray Ends 
                count = 0

        return ans 

a = [1,1,0,1,1,1,0,1,1]
sol = Solution()
ans = sol.findMaxConsecutiveOnes(a)
ans
```

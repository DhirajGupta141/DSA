##### Best time to buy and sell Stocks I : LC - 121  ( Only One Buy ans Sell Allowed)

```python
from typing import List

class Solution:
    def maxProfit(self, a: List[int]) -> int:

        buy = a[0]
        ans = 0
        for i in range(1, len(a)):
            buy = min(buy , a[i])  # finding the cheapest buy 
            ans = max( ans , a[i] - buy)  # what if I sell cheapest buy till now on today ? 

        return ans 

a = [7,1,5,8,3]
sol = Solution()
ans = sol.maxProfit(a)
ans  
```

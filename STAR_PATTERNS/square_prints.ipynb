```python
def pattern(n):

    try:

        for i in range(n):

            if i == 0 or i == n-1:
                print("* "*n, end = " ")
                print()
            else:
                print("*" + " "*(2*n-3) + "*" , end = " ")
                print()

    except Exception as e:
        raise e


n = 10
pattern(n)
```

#### Print Numbers In A Square Pattern 

#### Ascending ORDER PRINT 

```python
def printSquareNumber(n):
    try:
        ans = [[0]*n for i in range(n)]

        m = n  # n is row and m is column 

        top = 0 
        bottom = n - 1
        left = 0 
        right = m - 1

        num = 1 

        # print(f"Initialized Answer is : {ans}")

        while left <= right and top <= bottom:
            
            # move left to right 
            for i in range(left , right + 1):
                ans[top][i] = num 
                num  += 1
            
            top += 1

            # move top to bottom 

            for i in range(top , bottom + 1):
                ans[i][right] = num 
                num  += 1
            
            right -= 1


            
            # move right to left 
            for i in range(right , left - 1, -1):
                ans[bottom][i] = num 
                num  += 1
            
            bottom -= 1

            # move  bottom to top
            for i in range(bottom , top - 1, -1):
                ans[i][left] = num 
                num  += 1
            
            left += 1
        
        return ans 

    except Exception as e:
        raise e

n = 6
ans = printSquareNumber(n) 

for row in ans :
    print(row) 
```

##### Decending ORDER Number Print 

```python
def printSquareNumber(n):
    try:
        ans = [[0]*n for i in range(n)]

        m = n  # n is row and m is column 

        top = 0 
        bottom = n - 1
        left = 0 
        right = m - 1

        num = n*n 

        # print(f"Initialized Answer is : {ans}")

        while left <= right and top <= bottom:
            
            # move left to right 
            for i in range(left , right + 1):
                ans[top][i] = num 
                num  -= 1
            
            top += 1

            # move top to bottom 

            for i in range(top , bottom + 1):
                ans[i][right] = num 
                num  -= 1
            
            right -= 1


            
            # move right to left 
            for i in range(right , left - 1, -1):
                ans[bottom][i] = num 
                num  -= 1
            
            bottom -= 1

            # move  bottom to top
            for i in range(bottom , top - 1, -1):
                ans[i][left] = num 
                num  -= 1
            
            left += 1
        
        return ans 

    except Exception as e:
        raise e

n = 6
ans = printSquareNumber(n) 

for row in ans :
    print(row) 
```

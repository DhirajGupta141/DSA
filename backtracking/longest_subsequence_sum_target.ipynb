### Find the Longest Subset/Subsequence whose sum = Target in an array

TC : O(2^n)

SC : O(n) : Stack Size 

1. At each element hamare pass choice hai ki.. wo element answer ka part hoga ki nhi ..

2. Agar answer ka part h toh answer array me add karke.. index bada do.. else.. just index bada do ..

3. Base case : Jab i >= n 

4. Yaha backtracking ke Solution me hame func within func .. likhna hota hai.. and Java me jaise.. main se call krke.. Class me multiple methods hote the 

## Backtracking Solution 

```python
# Function Within Function for the backtracking solution 

def longestSubset(a, target):
    ans = []
    def backtrack(i,current_sum,sum_array):

        nonlocal ans  # making ans as nonlocal variable... and maintating the answer in this variable

        # **************** Base Case ************************
        if i>=len(a) or  current_sum > target:
            return
         
        if current_sum == target:
            if len(sum_array) > len(ans):
                # ans = sum_array.copy()
                ans = sum_array[:]
        

        # ********** hypothesis *******************

        # pick the current element for the answer 
        backtrack(i+1 , current_sum + a[i] , sum_array + [a[i]])

        # Dont pick the current element for the answer 
        backtrack(i+1 , current_sum , sum_array )


        # ************* Induction or Answer Calc ***************** 
        # we dont have anything returing from the backtrack function 

    backtrack(0,0,[])
    
    return ans
```

```python
if __name__ == '__main__':
    a = [4,3,8,2,1,7]
    target = 8
    ans = longestSubset(a,target)
    print(f"Longest Subset / Subsequence having sum of elements as target {target} is {ans}")
```

### Dynamic Programming Solution 

dp contains an entry for every sum from 0 to target:
 
TC : O(n * n * target) : Other n for the copy of the sequence in each of these iterations 

SC :  O(target*n) : Elements ka sum .. Target tak ham store kr rahe and hr ek sum me .. worst case me N elements ho skate h  

```python
def longestSubsetDP(a, target):
    try:
        n = len(a)
        dp = {0:[]}

        for num in a:

            # We need to iterate the snapshot of the DP .. which has been created till now... as we are updating dp variable at the bottom
            current_dp_snapshot = dp.copy()  

            for s, seq in current_dp_snapshot.items():
                new_sum = s + num

                if new_sum <= target:
                    new_seq = seq + [num]

                    if new_sum not in dp:
                        dp[new_sum] = new_seq
                    else:
                        if len(new_seq) > len(dp[new_sum]):
                            dp[new_sum] = new_seq

        return dp.get(target, []) , dp



    except Exception as e:
        print(f"Error in the DP Solution !")

if __name__ == '__main__':
    a = [4,3,8,2,1,7]
    target = 8
    ans , dp= longestSubsetDP(a,target)
    print(f"DP is : {dp}")
    print(f"Longest Subset / Subsequence having sum of elements as target {target} is {ans}")
```
